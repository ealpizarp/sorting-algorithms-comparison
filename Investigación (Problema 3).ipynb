* ¿Cómo se compara el algoritmo quicksort con el algoritmo mergesort en términos de rendimiento temporal?

## Propuesta resolutiva

### Hipotesis propuesta

La comparación de estos algoritmos teoricamente es un tanto limitante ya que el quickSort y el mergeSort presentan la misma complejidad temporal $\mathcal{O}(nlog(n))$ a excepción del peor caso en el quickSort. Para estos dos algoritmos de ordenamiento que funcionan bajo la metodología divide y conquista la forma en que podemos compararlos y determinar cual es más eficiente que el otro es mediante pruebas empíricas, en estas pruebas es posible  evidenciar cómo el quickSort tiene un mejor comportamiento temporal que el MergeSort en su caso promedio y mejor. El unico caso donde el MergeSort tiene una complejidad temporal más favorable es cuando se compara con el peor caso del quickSort.

### Procedimiento para demostrar
* Haremos una comparación empirica, vamos a analizar los casos mejor, promedio y peor al quicksort, pero también aunque sabemos que el mergesort no presenta casos vamos a ponerlo a prueba con los diferentes casos del quicksort, las pruebas se van realizar con los mismos datos.

# Algoritmos a utilizar
## Algoritmo Quicksort

```python
def quicksort(n):

    size = len(n)
    menores = []
    mayores = []

    if size == 1 or size == 0:
        return n

    else:
        pivote = n[(size - 1) // 2]

        for j in range(size):
            if n[j] < pivote:
                menores.append(n[j])
            elif n[j] > pivote:
                mayores.append(n[j])
        
        menores = quicksort(menores)
        mayores = quicksort(mayores)
        menores.append(pivote)

        return menores + mayores
```

## Algoritmo Merge sort

```python

def mergeSort(arreglo):

    if len(arreglo) <= 1:
        return arreglo
    else:
        mitad = int(len(arreglo)//2)

        izquierda = mergeSort(arreglo[:mitad])
        derecha = mergeSort(arreglo[mitad:])
    
        return merge(izquierda,derecha)
    

def merge(izquierda,derecha):
    i,j = 0,0
    result = []

    while i<len(izquierda) and j<len(derecha):

        if izquierda[i] <= derecha[j]:
            result.append(izquierda[i])
            i += 1
      
        else:
            result.append(derecha[j])
            j += 1

    result += izquierda[i:]
    result += derecha[j:]

    return result
```
