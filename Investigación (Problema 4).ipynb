### Divide  y Vencerás: 

#####  QuickSort:

```python
def quicksort(n):

    size = len(n)
    menores = []
    mayores = []

    if size == 1 or size == 0:
        return n

    else:
        pivote = n[(size - 1) // 2]

        for j in range(size):
            if n[j] < pivote:
                menores.append(n[j])
            elif n[j] > pivote:
                mayores.append(n[j])
        
        menores = quicksort(menores)
        mayores = quicksort(mayores)
        menores.append(pivote)

        return menores + mayores
```

### Intercambio Adyacente:
##### Insertion Sort:

```python
def insertionSort(arreglo): 
  
    for i in range(len(arreglo)): 
        
        j = i 
        
        while j > 0 and arreglo[j] < arreglo[j - 1]:
                temp = arreglo[j]
                arreglo[j] = arreglo[j - 1]
                arreglo[j-1] = temp
                j -= 1

    return arreglo
```

```python
from time import time

def medirTiempo(funcion, L):
    tiempoInicio = time()
    L = funcion(L)
    tiempoFinal = time()
    print("tiempo de ejecución de :",funcion.__name__,"Con: ", len(L), "elementos", tiempoFinal - tiempoInicio)
```
