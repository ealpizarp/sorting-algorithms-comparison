### Selection Sort

Con el selection sort tenemos que la complejidad temporal esta definida por:

$$
\begin{aligned}
\sum_{i=1}^{n-1} i = \frac{n^2 - n}{2} = \frac{1}{2} n^2 = \frac{1}{2} n
\end{aligned}
$$

Aplicando propiedades de $ \mathcal{O} $:

$$
\begin{aligned}
\therefore T(n) = \mathcal{O}(n^2)
\end{aligned}
$$

Debido a la naturaleza del algoritmo, tenemos que la complejidad temporal es la misma para todos los posibles casos, sin embargo, a la hora de hacer los experimentos vamos a tomar una lista ordenada ascendentemente como su mejor caso y para el peor caso tomaremos una lista ordenada descendentemente.

### Insertion Sort

Para el insertion sort tenemos que su peor caso es:

$$
\begin{aligned}
\sum_{i=1}^{n} \sum_{j=1}^{i} 1 = \sum_{i=1}^{n} i = \frac{n(n+1)}{2}
\end{aligned}
$$

Aplicando propiedades de $ \mathcal{O} $ tenemos que:

$$
\begin{aligned}
\therefore \underset{peor}{T(n)} = \mathcal{O}(n^2) \\
\end{aligned}
$$

Este caso se da cuando la configuración del arreglo de entrada es una permutación de elementos descendentemente ordenados. 

Para el caso promedio del insertion sort tenemos que su complejidad temporal esta definida por:

$$
\begin{aligned}
\sum_{i=1}^{n} \sum_{j=i+1}^{n}\frac{1}{2}  = \frac{n^2-n}{4} 
\end{aligned}
$$

De esta manera tenemos que aplicando propiedades de $ \mathcal{O} $

$$
\begin{aligned}
\therefore \underset{promedio}{T(n)} = \mathcal{O}(n^2) \\
\end{aligned}
$$

Para el caso donde el algoritmo tiene el mejor comportamiento, tenemos que la complejidad temporal es:

$$
\begin{aligned}
\sum_{i=1}^{n} 1 = n
\end{aligned}
$$

Aplicando las propiedades de $ \mathcal{O} $

$$
\begin{aligned}
\therefore \underset{mejor}{T(n)} = \mathcal{O}(n) \\
\end{aligned}
$$

Este caso se da cuando la configuración de entrada es un arreglo ascendentemente ordenado.


### Bubble Sort

Para el bubbleSort tenemos que su peor caso esta definido por:

$$
\begin{aligned}
\sum_{i=1}^{n-1} \sum_{j=1}^{i} 1 \\
\end{aligned}
$$


Con esto concluimos que:


$$
\begin{aligned}
\therefore \underset{peor}{T(n)} = \mathcal{O}(n^2) \\
\end{aligned}
$$

Este caso se da cuando la configuración de entrada es una permutación de elementos descendentemente ordenados.

De manera analoga tenemos que el caso promedio se define de la siguiente manera:

$$
\begin{aligned}
\sum_{i=1}^{n-1} \sum_{j=1}^{i} E[x] = \sum_{i=1}^{n-1} \sum_{j=1}^{i} \frac{1}{2} = \frac{(n-1)n}{4}
\end{aligned}
$$
De esta manera podemos concluir:

$$
\begin{aligned}
\therefore \underset{promedio}{T(n)} = \mathcal{O}(n^2)
\end{aligned}
$$

Este caso se da cuando la configuración de entrada esta dada por una permutacion de elementos aleatoria.

En el bubble sort tenemos que el mejor caso esta definido de esta manera:

$$
\begin{aligned}
\sum_{i=1}^{n-1} 1 = n - 1 \\
\end{aligned}
$$

Aplicando propiedades de $ \mathcal{O} $ tenemos:

$$
\begin{aligned}
\therefore \underset{mejor}{T(n)} = \mathcal{O}(n) \\
\end{aligned}
$$

Este caso se da cuando tenemos una configuración de entrada que es una permutacion de elementos ascendentemente ordenada.

### Quick Sort

Para el quicksort sabemos que su comportamiento es diferente ya que es un algoritmo basado en divide y conquista, el análisis de este algoritmo es un tanto distinto a los anteriores.

Para el mejor caso tenemos que se da cuando la selección del pivote es la mediana del arreglo de elementos que estemos ordenando, esto causa que la partición sea exactamente dos subsecuencias de tamaño $\frac{n}{2} $  cada una. Esto nos causa un arbol balanceado de altura $ \mathcal{h} $ donde se realizarán $\mathcal{2h} $ llamadas, lo cual va a ser mucho más favorable para la complejidad temporal del algorimo.

$$
\displaystyle
\therefore \underset{mejor}{T(n)} = \mathcal{O}(n log(n))
$$


Con respecto al caso promedio tenemos que la probabilidad en donde la selección del pivote sea lo suficientemente buena para particionar en dos subsecuencias de $\frac{1}{4}n $ o $\frac{3}{4}n $ es de $\frac{1}{2}$. Desarollando el análisis respectivo a este caso tendriamos que la complejidad temporal es de:


$$
\displaystyle
\therefore \underset{promedio}{T(n)} = \mathcal{O}(n log(n))
$$

Ahora, con respecto al peor caso, este se da cuando la selección del pivote sea el máximo o el mínimo elemento de la lista, esto nos causaria que la partición resulte en dos subsecuencias donde una tiene un tamáno de $ n-1 $ y otra tenga una tamaño de $ 0 $ lo cual causaria un arbol de llamadas muy desbalanceado. Con esto podemos concluir que:

$$
\displaystyle
\therefore \underset{peor}{T(n)} = \mathcal{O}(n^2)
$$

### Merge Sort 

El mergesort es un algoritmo completamente basado en la tecnica de diseño de algoritmos divide y conquista. Debido a que el procedimiento del mergesort para ordenar una permutacion de elementos siempre va a ser el mismo, tenemos que no muestra casos distintos. Para este algoritmo tenemos que su patron de recurrencia será el siguiente


$$
\displaystyle
\ T(n) = 2T\big(\frac{n}{2}\big) + n
$$


Si desarollamos un poco este patron vamos a tener de manera generalizada:

$$
\displaystyle
\ T(n) = n + nlog(n)
$$


Aplicando propiedades de $ \mathcal{O} $, tenemos que:


$$
\displaystyle
\therefore T(n) = \mathcal{O}(nlog(n))
$$


Entonces podemos conlcuir que para todos los posibles casos vamos a tener $ nlog(n) $ como complejidad temporal, sin embargo, para efectos de prueba vamos a tomar configuraciones de entrada que sean permutaciones ascendentemente ordenados para el mejor caso, aleatorias para el caso promedio y descendentemente ordenadas para el peor caso. 




## Algoritmos a utilizar


A continuación vamos a mostrar cuales son los algoritmos que vamos a utilizar para realizar los experimentos. Es importante notar que estos algoritmos los hicimos de tal manera que fueran más claros de entender y que tuvieran la complejidad temporal esperada, sin embargo, la complejidad espacial en algunos de estos algorimos no es la más optima. 


### Selection Sort

````python
def selectionSort(arreglo):
  n = len(arreglo)
  for i in range(n):
    minimo = i

    for j in range(i+1, n):
      if (arreglo[j] < arreglo[minimo]):
        minimo = j

    temp = arreglo[i]
    arreglo[i] = arreglo[minimo]
    arreglo[minimo] = temp
    
  return arreglo
````

### Insertion Sort

````python
def insertionSort(arreglo): 
  
    for i in range(len(arreglo)): 
        
        j = i 
        
        while j > 0 and arreglo[j] < arreglo[j - 1]:
                temp = arreglo[j]
                arreglo[j] = arreglo[j - 1]
                arreglo[j-1] = temp
                j -= 1

    return arreglo
````

### Bubble Sort

````python
def bubbleSort(arreglo):
    n = len(arreglo)
    i = n-1
    while  i > 0:
        for j in range(0, i):
            if arreglo[j] > arreglo[j+1] :
                arreglo[j], arreglo[j+1] = arreglo[j+1], arreglo[j]
        i -= 1
    return arreglo
````

### Quick Sort

````python
def quicksort(n):

    size = len(n)
    menores = []
    mayores = []

    if size == 1 or size == 0:
        return n

    else:
        pivote = n[(size - 1) // 2]

        for j in range(size):
            if n[j] < pivote:
                menores.append(n[j])
            elif n[j] > pivote:
                mayores.append(n[j])
        
        menores = quicksort(menores)
        mayores = quicksort(mayores)
        menores.append(pivote)

        return menores + mayores
````

### Merge Sort

````python
def mergeSort(arreglo):

    if len(arreglo) <= 1:
        return arreglo
    else:
        mitad = int(len(arreglo)//2)

        izquierda = mergeSort(arreglo[:mitad])
        derecha = mergeSort(arreglo[mitad:])
    
        return merge(izquierda,derecha)
    

def merge(izquierda,derecha):
    i,j = 0,0
    result = []

    while i<len(izquierda) and j<len(derecha):

        if izquierda[i] <= derecha[j]:
            result.append(izquierda[i])
            i += 1
      
        else:
            result.append(derecha[j])
            j += 1

    result += izquierda[i:]
    result += derecha[j:]

    return result
````

## Conclusiones

<br/>

Ahora que tenemos los resultados de los experimentos podemos referirnos a la resolución del problema inicial:

* Para cada algoritmo de ordenamiento estudiado hasta el momento en el curso —a saber: ordenamiento por selección, ordenamiento por inserción, ordenamiento burbuja, quicksort, y mergesort—, si ejecutamos el algoritmo en una máquina concreta, ¿coincide su comportamiento temporal con los resultados predichos teóricamente (incluyendo los distintos casos)?

En primera instancia basandonos en los resultados de las pruebas empíricas podemos concluir con certeza que la complejidad temporal no va a ser la misma para distintas máquinas concretas ni tampoco predice el comportamiento exacto de un algoritmo, sin embargo, indudablemente la teoría coincide con el comportamiento general que un algoritmo va tener a la hora de ejecución.

En los primeros tres experimentos tuvimos que al realizar pruebas empíricas en diferentes puntos de crecimiento según el caso respectivo del experimento casi todos los resultados coincidieron con el comportamiento teórico esperado y la aproximación fue lo suficientemente buena para tener una idea de como se comportaría el algoritmo. Esto lo verificamos con la teoría de regresión lineal y el coeficiente de determinación. La verificación de dichos datos se puede hacer con la tabla de resultados de cada experimento. 

<br/>

![Regresión Lineal](https://upload.wikimedia.org/wikipedia/commons/3/3a/Linear_regression.svg "Regresión Lineal")

<br/>

En algunas pruebas dichos resultados fueron sorprendentes, debido a que la aproximación teórica fue prácticamente igual al resultado empírico. El insertion sort y el selection sort fueron unos de los que mostraron una relación muy estrecha entre la proyección teórica y datos empíricos. 

Los resultados que obtuvimos del experimento cuatro ayudaron a reforzar más el argumento expuesto en nuestra hipótesis, en la graficación de estos mismos podemos observar una vez más la estrecha relación de la teoría con el comportamiento empírico tomando una muestra de `0 - 10000` elementos y logrando obtener una idea más general de cada algoritmo en su caso promedio. 

Basado en los argumentos expuestos anteriormente, llegamos concluir con certeza:
<br/>

* El comportamiento temporal de un algoritmo en una maquina concreta no se puede estimar exactamente con los resultados predichos teóricamente.
<br/>

* Los tiempos de ejecución de un algoritmo pueden variar según multiples factores ya sean los recursos de una máquina, procesos del SO, temperatura, entre otros.
<br/>

* La proyección teórica indudablemente tiene una estrecha relación con los resultados empíricos ya que en la mayoría de los casos logra ser muy precisa esto quiere decir que la teoría sí coincide en todos los casos.
<br/>

* El crecimiento de un algoritmo puede ser proyectado según el análisis teorico y las propiedadees de  $ \mathcal{O} $.
